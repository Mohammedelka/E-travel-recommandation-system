```python
def min_distance_between_words(comment1, comment2, word2vec_model):
    # Split the comments into words
    words1 = comment1.split()
    words2 = comment2.split()

    # Calculate the min distance between words
    min_distance = min(word2vec_model.wv.vocab.get(w, 0) for w in words1 + words2)
    return min_distance
```
